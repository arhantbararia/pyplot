```python
from matplotlib import pyplot as plt
```

```python
##basic start
plt.title("Pie Chart")
```

```python
slices = [60 ,40]
plt.pie(slices) ###.pie
```

```python
slices = [60 ,40]
labels = ["sixty" , "forty"] ##respectively to slices 
plt.pie(slices , labels = labels) ###.pie
```

```python
## to add distinction wedge
plt.pie(slices , labels = labels , wedgeprops = {"edgecolor" : 'black'}) ##this is basic, look up documentation for more
```

```python
#custom color
colors = ["Red" , "Yellow"] ##respectively to slices
plt.pie(slices , labels = labels ,colors = colors , wedgeprops = {"edgecolor" : 'black'})
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
plt.pie(slices , labels = labels , wedgeprops = {"edgecolor" : 'black'})
```

```python
# for putting emphasis on a single data set , we use explode parameter of pie method


explode = [0 , 0 , 0.1 ,0, 0]  ##0.1 shows 10% of Radius away from the center 

plt.pie(slices , labels = labels ,explode = explode, wedgeprops = {"edgecolor" : 'black'})
```

```python
# for shadow 

plt.pie(slices , labels = labels , wedgeprops = {"edgecolor" : 'black'} , shadow = True)
```

```python
## for setting starting angle from horizontal

plt.pie(slices , labels = labels , wedgeprops = {"edgecolor" : 'black'} , shadow = True , startangle = 90)
```

```python
## to label each data set with percentages 

plt.pie(slices , labels = labels , wedgeprops = {"edgecolor" : 'black'} , autopct = '%1.1f%%') ##
```
