```python
import pandas as pd

data = pd.read_csv('data.csv')
ids= data["Responder_id"]
language_responses = data["LanguagesWorkedWith"]
```

```python
from collections import Counter

language_counter = Counter()
```

```python
for responses in language_responses:
    language_counter.update(responses.split(";"))
```

```python
print(language_counter)
```

```python
print(language_counter.most_common(30))   ##gives out top 30 most common programming languages in touplt form
```

```python
language = []
popularity= []

for element in language_counter.most_common(30):
    language.append(element[0])
    popularity.append(element[1])
```

```python
language.reverse()
```

```python
plt.barh(language , popularity)
```
