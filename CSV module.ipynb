```python
with open('names.csv' , 'r') as csv_file:    #opening a file and iterating through it till we reach the end
    csv_reader = csv.reader(csv_file)

    for line in csv_reader:                  ##here note that for loop is opened inside with open file
        print(line)
```

```python
## making a new file
```

```python
with open('names.csv' , 'r') as csv_file:   
    csv_reader = csv.reader(csv_file)
    
    with open("new_csvfile.csv" , "w") as new_file:    ##
        csv_writer = csv.writer(new_file, delimiter = "-")
    
        for line in csv_reader:
            csv_writer.writerow(line)      ##note for is inside newfile which is further inside csv_file
```

```python
with open('new_csvfile.csv' , 'r') as csv_file:   ##opening our newly created file
    csv_reader = csv.reader(csv_file)   ## this shows you need to add delimiter parameter for any delimiter other than ","
    
    for line in csv_reader:
        print(line)
```

```python
with open('new_csvfile.csv' , 'r') as csv_file:   ##opening our newly created file
    csv_reader = csv.reader(csv_file , delimiter = "-")   ## this shows you need to add delimiter parameter for any delimiter other than ","
    
    for line in csv_reader:
        print(line)
```

```python
##using DictReader for better indexing by keys
```

```python
with open('names.csv' , 'r') as csv_file:   
    csv_reader = csv.DictReader(csv_file)
    
    for line in csv_reader:
        print(line)                  ##note first_name , last_name , email are keys 
```

```python
with open('names.csv' , 'r') as csv_file:   
    csv_reader = csv.DictReader(csv_file)
    
    for line in csv_reader: 
        print(line["email"])         ##bettter indexing by keys
```

```python
with open('names.csv' , 'r') as csv_file:   
    csv_reader = csv.DictReader(csv_file)
    
    with open("new_csvfile.csv" , "w") as new_file:    
        fieldnames = ["first_name" , "last_name"]   #if you don't want to include certain field/keys  in some file
        csv_writer = csv.DictWriter(new_file, fieldnames = fieldnames , delimiter = "-")
        
        csv_writer.writeheader()            #write keys 
        
        for line in csv_reader:
            del line['email']                ##if you don't want to include certain field/keys in some file
            csv_writer.writerow(line) 
```
