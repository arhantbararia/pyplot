```python
plt.title("Median Salary (USD)  by age")
plt.plot(ages_x , dev_y )
plt.plot(ages_x , py_dev_y)
plt.fill_between(ages_x , py_dev_y , alpha = 0.25) ###alpha sets the opacity 
plt.xlabel = "age"
plt.ylabel = "Median Salary (USD)"
plt.legend(['All Devs' , 'Python Devs'])##
plt.show()
```

```python
overall_median = 52787 ##setting a bottom minimum 

plt.title("Median Salary (USD)  by age")
plt.plot(ages_x , dev_y )
plt.plot(ages_x , py_dev_y)
plt.fill_between(ages_x , py_dev_y ,overall_median ,  alpha = 0.25) ## 3rd variable of median 
plt.xlabel = "age"
plt.ylabel = "Median Salary (USD)"
plt.legend(['All Devs' , 'Python Devs'])##
plt.show()
```

```python
#using js_dev_y to set minimum 

plt.title("Median Salary (USD)  by age")
plt.plot(ages_x , dev_y )
plt.plot(ages_x , py_dev_y)
plt.fill_between(ages_x, py_dev_y, js_dev_y,
                 where=(py_dev_y > js_dev_y), ####
                 interpolate=True, alpha=0.25, label='Above Avg')
plt.xlabel = "age"
plt.ylabel = "Median Salary (USD)"
plt.legend(['All Devs' , 'Python Devs'])###
plt.show()
```
